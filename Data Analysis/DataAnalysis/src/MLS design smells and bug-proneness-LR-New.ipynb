```python
def fishers_exact_test (data):
    odds_ratio, p_value = stats.fisher_exact(data, alternative='two-sided')
    return odds_ratio, p_value
```

```python
import decimal

# create a new context for this task
ctx = decimal.Context()

# 20 digits should be enough for everyone :D
ctx.prec = 10

def float_to_str(f):
    """
    Convert the given float to a string,
    without resorting to scientific notation
    """
    d1 = ctx.create_decimal(repr(f))
    return format(d1, 'f')
```

```python
smell_list = selected_columns[6:21]
# print(smell_list)

for sys in sys_names:
    sys_name = sys
    report_path = os.path.join(data_path_root, 'data','fishertest', sys_name + '_fisher-s-new.csv')
    sys_input_data=merged_data[merged_data['system']==sys_name]
    release_list = list(set(sys_input_data['version'].values))
    # print(release_list)
    # print(sys_input_data.head(2))
    # print(sys_input_data.head(2))
    header = "System,Release,Smell,S_B,NS_B,S_NB,NS_NB,Odds_ratio,pvalue"
    report_txt=header + '\n'
    for rl in release_list:
        print('System {} Release {}'.format(sys_name, rl))
        print('==============================')
        rel_df = sys_input_data[sys_input_data['version']==rl] # for selected release
#         jni_df = rel_df. query('JNINativeFlag ==1 & Smelly==1') # for JNI files in selected release
#         n_of_jni_files = jni_df.shape[0] 
        for sml in smell_list:
#             smelly_buggy = jni_df[(jni_df[sml]>0) & (jni_df['InducingFlag'])==1].shape[0]
#     #         print(smelly_buggy)
#             non_smelly_buggy = jni_df[(jni_df[sml]==0) & (jni_df['InducingFlag'])==1].shape[0]
#     #         print(non_smelly_buggy)
#             smelly_non_buggy = jni_df[(jni_df[sml]>0) & (jni_df['InducingFlag'])==0].shape[0]
#             non_smelly_non_buggy = jni_df[(jni_df[sml]==0) & (jni_df['InducingFlag'])==0].shape[0]

                    
            sml_df = rel_df.query ('Smelly==1') 
    
            qry_str1 = sml +'>0 & InducingFlag==1'
            s_b_df = sml_df.query(qry_str1)
            smelly_buggy = s_b_df.shape[0]
        
            qry_str2 = sml +'==0 & InducingFlag==1'
            ns_b_df = sml_df.query(qry_str2)
            non_smelly_buggy = ns_b_df.shape[0]

            qry_str3 = sml +'>0 & InducingFlag==0'
            s_nb_df = sml_df.query(qry_str3)
            smelly_non_buggy = s_nb_df.shape[0]

            qry_str4 = sml +'==0 & InducingFlag==0'
            ns_nb_df = sml_df.query(qry_str4)
            non_smelly_non_buggy = ns_nb_df.shape[0]
            
            test_data = [[smelly_buggy, non_smelly_buggy],
                         [smelly_non_buggy, non_smelly_non_buggy]]
            odd_r, p_val = fishers_exact_test(test_data)
            txt_line = "{},{},{},{},{},{},{},{},{}".format(sys_name, rl, sml, smelly_buggy, non_smelly_buggy,
                                            smelly_non_buggy, non_smelly_non_buggy, odd_r, float_to_str(p_val))
            print(txt_line)
            report_txt = report_txt + txt_line + '\n'
    report_file = open(report_path,"w")
    report_file.write(report_txt)
    report_file.close()
```
