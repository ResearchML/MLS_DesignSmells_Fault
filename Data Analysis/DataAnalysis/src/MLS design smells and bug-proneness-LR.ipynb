```python
def load_csv(path):
    data = pd.read_csv(path, encoding = 'cp1251')
    return data
```

```python
def mmw_test(d1, d2):
    st, p_value = stats.mannwhitneyu(d1, d2, alternative= 'two-sided')
    return st, p_value
```

```python
selected_columns = ['system', 'version', 'release', 'FileName', 'JNINativeFlag', 'Smelly',
       'excessiveInterlangCommunication','Toomuchclustring', 'ToomuchScattering', 'UnusedMethodDeclaration',
       'UnusedMethodImplementation', 'UnusedParameter',#'AssumingSafeReturnValue', #'ExcessiveObjects'
                     'NotHandlingExceptions' 
                    #,'NotCachingObjects'
        ,'NotSecuringLibraries', 'HardCodingLibraries','NotUsingRelativePath', 'MemoryManagementMismatch',
       'LocalReferencesAbuse', 'CLOC', 'PrevFixing', 'CodeChurn', 'InducingFlag'] 
```

```python
columns_to_export = ['system', 'version', 'release', 'FileName', 'JNINativeFlag','Smelly',
       'excessiveInterlangCommunication','Toomuchclustring', 'ToomuchScattering', 'UnusedMethodDeclaration',
       'UnusedMethodImplementation', 'UnusedParameter', 'AssumingSafeReturnValue', 'ExcessiveObjects',
                     'NotHandlingExceptions' 
                    ,'NotCachingObjects'
        ,'NotSecuringLibraries', 'HardCodingLibraries','NotUsingRelativePath', 'MemoryManagementMismatch',
       'LocalReferencesAbuse', 'CLOC', 'PrevFixing', 'CodeChurn', 'InducingFlag']
```

```python
merged_data = pd.DataFrame([], columns = selected_columns)
for i in range (len(sys_names)):
    sys_name = sys_names[i]
    detail_data_path = os.path.join(data_path_root, 'data', sys_name + '_details_cleaned.csv')
    data = load_csv(detail_data_path)
    data['CodeChurn'] = data['Added'] + data['Removed'] 
    data = data.drop(['Added','Removed'], axis=1)
    data_selected = data[columns_to_export]
    if i==0:
        merged_data = data_selected
    else:
        merged_data = merged_data.append(data_selected)
    print('data loaded from:{}'.format(detail_data_path))
    print ('current rows:{}'.format(merged_data.shape[0]))
```
