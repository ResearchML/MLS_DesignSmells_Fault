```python
import os
import pandas as pd
import numpy as np
from scipy import stats
from pathlib import Path
```

```python
 path_root = os.getcwd()
 data_path_root = Path(path_root).parent
```

```python
def load_csv(path):
    data = pd.read_csv(path)
    return data
```

```python
sys_name = 'javacpp'
fpath = os.path.join(data_path_root,'data','plotdata',sys_name+'smells.csv')
print(fpath)
data_df = load_csv(fpath)
print (data_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
cols = data_df.columns
cols = cols[1:]
fig = plt.figure(figsize=(12,7))
for i in range(len(cols)):
    plt.plot(data_df[cols[i]], label=cols[i])

# # Add legend
plt.legend(loc='upper center', )
# Add title and x, y labels
# plt.title("Random Walk Example", fontsize=16, fontweight='bold')
# plt.suptitle("Random Walk Suptitle", fontsize=10)
plt.xlabel("Releases")
plt.ylabel("Number of Files with Multi-language Design Smells")
plt.show()
```
