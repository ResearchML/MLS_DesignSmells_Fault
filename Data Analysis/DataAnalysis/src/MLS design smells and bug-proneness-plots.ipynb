```python
import os
import pandas as pd
import numpy as np
from scipy import stats
from pathlib import Path
```

```python
 path_root = os.getcwd()
 data_path_root = Path(path_root).parent
```

```python
def load_csv(path):
    data = pd.read_csv(path,delimiter=';')
    return data
```

```python
sys_name = 'java-smt'
fpath = os.path.join(data_path_root,'data','plotdata',sys_name+'.csv')
fig_path = os.path.join(data_path_root,'data','figures',sys_name+'.pdf')
print(fpath)
data_df = load_csv(fpath)
print (data_df)
```

```python
data_df = data_df.loc[:, (data_df != 0).any(axis=0)]
```

```python
# import seaborn as sns
import matplotlib.pyplot as plt 
cols = data_df.columns
cols = cols[1:]
release = data_df.shape[0]
fig = plt.figure(figsize=(15,10))
markers=['o','v','^','<','>','s','*','h','H','+','x','X','p','P','D','d','|','_','1','2','3','4','8',',','.']
for i in range(len(cols)):
    plt.plot(data_df[cols[i]], label=cols[i], marker=markers[i], markersize =8, linewidth =2.5)

# # Add legend
plt.legend(loc='center right', ncol = 1, fontsize = 'large')
ax = plt.gca()
ax.set_facecolor((0.97, 0.97, 0.97))

# Add title and x, y labels
# plt.title("Random Walk Example", fontsize=16, fontweight='bold')
# plt.suptitle("Random Walk Suptitle", fontsize=10)
plt.xlabel("Releases", fontsize = 'x-large', labelpad=8)
plt.ylabel("Number of Files with Multi-language Design Smells", fontsize = 'x-large', labelpad=15)
fig1 = plt.gcf()
# print(np.arange(1, 11, step=1))
plt.xticks(np.arange(1, release, step=1))
plt.show()
plt.draw()
fig.savefig(fig_path, pad_inches=None)
```
